```python
import torch 
import numpy as np
import pandas as pd
import torch.nn as nn
import torch.nn.functional as F
from sklearn.model_selection import train_test_split
```

```python
df=pd.read_csv('music_train.csv',index_col='id')
df=df.fillna(0)
df
```

```python
x=(df.iloc[:,1:24]) 
x['topic']=df['topic']
x
 
```

```python
 y=pd.DataFrame(df['genre'])
 y=y.squeeze(1)
 y
```

```python
x=torch.tensor(x.values, dtype=torch.float32)
y=torch.tensor(y.values)
x,y
```

```python
val_percent = 0.05
input_size = 24
hidden_size = 13
learning_rate = 0.001
num_epochs = 1000
num_classes = 7
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(x, y, test_size=val_percent, random_state=42)
```

```python
class NeuralNet(nn.Module):
    def __init__(self, input_size, hidden_size, num_classes):
        super(NeuralNet, self).__init__()
        self.fc1 = nn.Linear(input_size, hidden_size)
        self.fc2 = nn.Linear(hidden_size, hidden_size)
        self.fc3 =  nn.Linear(hidden_size, hidden_size)
        self.fc4 =  nn.Linear(hidden_size, num_classes)

    
    def forward(self, x):
        out = self.fc1(x)
        out = F.relu(out)
        out = self.fc2(out)
        out = F.relu(out)
        out = self.fc3(out)
        out = self.fc4(out)
        return out


model = NeuralNet(input_size, hidden_size, num_classes)
model
```

```python
criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.parameters(), lr=learning_rate)
```

```python
for epoch in range(num_epochs): 
    X = torch.FloatTensor(X_train)
    Y = torch.LongTensor(Y_train)
    out = model(X)
    loss = criterion(out, Y)
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
     
    acc = 100 * torch.sum(Y==torch.max(out.data, 1)[1]).double() / len(Y)
    print ('Epoch [{}/{}] , Loss: {:.4f} , Accuracy: {:.4f}' .format(epoch+1, num_epochs , loss.item(), acc.item() ))
    
```

```python
X_test = torch.FloatTensor(X_test)
Y_test = torch.LongTensor(Y_test)
out = model(X_test)
predicted = torch.max(out.data, 1)[1]
print('Accuracy of the network %.4f %%' % (100 * torch.sum(Y_test==predicted).double() / len(Y_test)))
```
